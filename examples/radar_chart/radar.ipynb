# Examples for Radar Chart

* This jupyter notebook will contain code snippets, which will explain how can you make use of `soccerplots` to make radar charts.

* The code for `soccerplots` is made using `Numpy==1.18.5` and `Matplotlib==3.3.0`. 

## Documentation for plot_radar:

* We will be using `plot_radar` function to plot our radar charts.

* The functions looks like this:
    
   ```python
   plot_radar(ranges, params, values, radar_color, label_fontsize=10, range_fontsize=6.5, figax=None, filename=None, dpi=300, title=dict())
   ```


* And here is the description of each parameter:
   
   * ranges -- list of tuples containing min and max value for each parameter.
   
   * params -- list of value containing parameter names.
   
   * values -- list containing the value for each parameter.
   
   * radar_color -- list containing two color values.
   
   * label_fontsize -- fontsize for your labels. Default=10.
   
   * range_fontsize -- fontsize for the range values. Default=6.5.
   
   * figax -- tuple containing figure and axis object. Default=None.
   
   * filename -- the name per which the file will be saved. Default=None.
   
   * dpi -- dots per inch value. Default=300.
   
   * title -- dict containing information of title and subtitle.
   
* **Note**: You will need minimum of 3 parameters to make the radar-chart.   
   
* Let's see some code-snippets.

````python
## here are all the parameter names
params = ["xG90", "Shots", "Touches In Box", "xGChain", "Passing%"]

## here is the min and max range values for each parameter
ranges = [(0.16, 0.60), (1.4, 3.9), (1.2, 4.0), (0.39, 1.15), (63, 83)]

## here is the value for each parameter
values = [0.91, 3.07, 6.13, 0.77, 68]

## plot the radar chart
fig, ax = radar_chart.plot_radar(ranges=ranges, params=params, values=values, radar_color=['#CE3B2C', '#F4CF39'])
````
