```python
from mplsoccer.pitch import Pitch
import matplotlib.pyplot as plt
```

```python
figsize = (11,8)
fig_aspect = figsize[0]/figsize[1]
fig = plt.figure(figsize=figsize)
# horizontal full
horizontal_full = Pitch(orientation='horizontal', view='full')
horizontal_full_height = 0.45
horizontal_full_width = horizontal_full_height*horizontal_full.ax_aspect/fig_aspect
ax_horizontal_full = fig.add_axes((0.05, 0.05, horizontal_full_width, horizontal_full_height))
horizontal_full.draw(ax=ax_horizontal_full)
horizontal_full.annotate("Horizontal Full",(85,40), va='center', ha='center', ax=ax_horizontal_full)
# vertical full
vertical_full = Pitch(orientation='vertical', view='full')
vertical_full_height = 0.85
vertical_full_width = vertical_full_height*vertical_full.ax_aspect/fig_aspect
ax_vertical_full = fig.add_axes((0.1 + horizontal_full_width, 0.05, vertical_full_width, vertical_full_height))
vertical_full.draw(ax=ax_vertical_full)
vertical_full.annotate("Vertical Full",(85,40), va='center', ha='center', ax=ax_vertical_full)
# horizontal half
horizontal_half = Pitch(orientation='horizontal', view='half')
horizontal_half_width = 0.2
horizontal_half_height = horizontal_half_width/horizontal_half.ax_aspect*fig_aspect
ax_horizontal_half = fig.add_axes((0.05, 0.1 + horizontal_full_height, horizontal_half_width, horizontal_half_height))
horizontal_half.draw(ax=ax_horizontal_half)
horizontal_half.annotate("Horizontal half",(85,40), va='center', ha='center', ax=ax_horizontal_half)
# vertical half
vertical_half = Pitch(orientation='vertical', view='half')
vertical_half_width = 0.25
vertical_half_height = vertical_half_width/vertical_half.ax_aspect*fig_aspect
ax_vertical_half = fig.add_axes(((0.05 + horizontal_full_width)-vertical_half_width, 0.1 + horizontal_full_height,
                                 vertical_half_width, vertical_half_height))
vertical_half.draw(ax=ax_vertical_half)
vertical_half.annotate("Vertical half",(85,40), va='center', ha='center', ax=ax_vertical_half);
```
